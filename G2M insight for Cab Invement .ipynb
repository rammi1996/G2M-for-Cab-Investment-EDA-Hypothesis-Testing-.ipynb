Hypothesis Testing 


Hyposthesis 1: IS there any difference in Profit based on Gender

H0: There is no diffrence in profit regrading Gender.
H1: There is diffrence in profit regarding Gender.

```python
Companies=['Pink Cab ', 'Yellow Cab']
```

```python
# Assume df is your DataFrame with 'Gender', 'Company', and 'Profit' columns
companies = ['Pink Cab', 'Yellow Cab']

for company in companies:
    print(f"\n{company}:")

    # Filter data by Gender and Company
    a = df[(df['Gender'] == 'Male') & (df['Company'] == company)]['Profit']
    b = df[(df['Gender'] == 'Female') & (df['Company'] == company)]['Profit']

    # Output the size of each group
    print(f"size a: {a.shape[0]} , size b: {b.shape[0]}")

    # Re-run the hypothesis test if both groups have data
    if a.shape[0] > 0 and b.shape[0] > 0:
        t_stat, p_value = stats.ttest_ind(a.values, b.values)
        print(f"t_statistics is {t_stat}")
        print(f"p_value is {p_value}")

        if p_value < 0.05:  # Assuming alpha = 0.05
            print(f'We accept the alternative hypothesis (H1) that there is a difference regarding profit by gender for {company}')
        else:
            print(f'We accept the null hypothesis (H0) that there is no difference regarding profit by gender for {company}')
    else:
        print(f"Insufficient data for {company}")
```
